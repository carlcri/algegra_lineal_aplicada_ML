# Las tres Transformaciones

Partamos del concepto que hemos manejado de las matrices como transformaciones lineales. Partiendo de la descomposicion en vectores y valores propios, o valores y vectores singulares, es posible pensar que cada parte de estas submatrices, por decirlo de alguna manera, son a su vez transormaciones lineales.

Vamos a estudiar los efectos de cada una de estas matrices sobre el ciruculo unitario ¿Porque el circulo Unitario?
Porque va a resultar mas facil ves sus efectos.

Empezemos por dibujar un circulo unitario

```python

x = np.linspace(-1,1,2100)
y = np.sqrt(1-x**2)
y_neg = -y

plt.plot(x, y, color='green')
plt.plot(x,y_neg, color='blue')
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```

Ahora iremos modificando el codigo anterior, para poder obtener el ciruculo pero transformado. Recuerda solo va a fucnionar para una matriz A de 2x2.

```python
#A = np.identity(2)
A = np.array([[3,7],[5,2]])
print(A)

# Ciruculo Unitario
x = np.linspace(-1,1,2100)
y = np.sqrt(1-x**2)
y_neg = -y

# Ciruculo Unitario Transformado -- parte positiva
x1 = A[0,0]*x + A[0,1]*y
y1 = A[1,0]*x + A[1,1]*y


plt.plot(x1, y1, color='green')

plt.xlim(-8.5,8.5)
plt.ylim(-8.5,8.5)
plt.show()
```

¡Muy bien! Hemos graficado una parte del ciruculo transformado, que tal si lo graficamos completo.


```python
#A = np.identity(2)
A = np.array([[3,7],[5,2]])

# Ciruculo Unitario
x = np.linspace(-1,1,2100)
y = np.sqrt(1-x**2)
y_neg = -y

# Ciruculo Unitario Transformado -- parte positiva
x1 = A[0,0]*x + A[0,1]*y
y1 = A[1,0]*x + A[1,1]*y

# Circulo Unitatario Transoformado -- parte negativa
x1_neg = A[0,0]*x + A[0,1]*y_neg
y1_neg = A[1,0]*x + A[1,1]*y_neg

#Grafica Circulo Untario Transformado
plt.plot(x1, y1, color='green')
plt.plot(x1_neg, y1_neg, color='red')
plt.xlim(-8.5,8.5)
plt.ylim(-8.5,8.5)
plt.show()
```

Vamos añadir al codigo anterior, los vectores $i$ y $j$ transformados

```python
A = np.array([[3,7],[5,2]])

# Vectores 
i_vector = np.array([[1,0]]).T
j_vector = np.array([[0,1]]).T

# Vectores Transformados

i_vector = A.dot(i_vector)
j_vector = A.dot(j_vector)

print(i_vector)
print(j_vector)
```

¿Notas algo familiar en los Vectores Transformados? Si, que el vector $i$ transformado es igual a la primera columna de A, y $j$ transformado es igual a la segunda columna de A. Veamos si podemos simplificar la implementacion

```python
A = np.array([[3,7],[5,2]])

i_vector = A[:,0]
j_vector = A[:,1]

print(i_vector)
print(j_vector)
```
